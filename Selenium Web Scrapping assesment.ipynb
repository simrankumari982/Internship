Q2: Write a python program to scrape data for “Data Scientist” Job position in
“Bangalore” location. You have to scrape the job-title, job-location,
company_name, full job-description. You have to scrape first 10 jobs data.

```python
driver.close()
driver=webdriver.Chrome("C:/Users/Simran Kumari/Downloads/chromedriver_win32/chromedriver.exe")
driver.get("https://www.naukri.com/")
skills=driver.find_element_by_id("qsb-keyword-sugg")
skills.send_keys("Data Scientist")
loc=driver.find_element_by_id("qsb-location-sugg")
loc.send_keys("Bangalore")
button=driver.find_element_by_class_name("btn")
button.click()
driver.get(driver.current_url)
jobs=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
locations=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
companies=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
com=[]
location_s=[]
for company in companies:
    com.append(company.text)
for location in locations:
    location_s.append(location.text)
urls=[]
job_title=[]
job_desc=[]
for job in jobs:
    urls.append(job.get_attribute('href'))
    job_title.append(job.get_attribute('title'))
for url in urls:
    driver.get(url)
    try:
        job_desc.append(driver.find_element_by_xpath("//div[@class='detail dang-inner-html']").text.replace("\n"," "))
    except:
        job_desc.append(driver.find_element_by_xpath("//div[@class='clearboth description']").text.replace("\n"," "))
data=[]
for i in range(0,10):
    data.append(
    {
        'Job Title':job_title[i],
        'Company':com[i],
        'Location':location_s[i],
        'Job Description':job_desc[i]
    })
df=pd.DataFrame(data)
df
```
