1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
url="https://en.wikipedia.org/wiki/Main_Page"
page=requests.get(url)
soup=bs(page.content,'html.parser')
headers=soup.find_all(['h1','h2','h3','h4','h5','h6'])
headers_result=[]
for header in headers:
    headers_result.append(header.get_text().replace("\n",""))
df=pd.DataFrame(data=headers_result)
df
```

2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
url="https://www.imdb.com/chart/top/"
page=requests.get(url)
soup=bs(page.content,'html.parser')
movies=soup.find_all('td',class_='titleColumn')
imdb_ratings=soup.find_all('td',class_='ratingColumn imdbRating')
ratings=[]
title=[]
rank=[]
yr=[]
i=0
for rating in imdb_ratings:
    ratings.append(rating.get_text().replace("\n",""))
for movie in movies:
    rank.append(i+1)
    title.append(movie.find("a").get_text())
    y=movie.find("span").get_text().replace("(","")
    y=y.replace(")","")
    yr.append(y)
    i=i+1
data={'Rank':rank,'Name':title,'IMD Ratings':ratings,'Year of release':yr}
df=pd.DataFrame(data)
result=df.head(100)
```

```python
result=result.set_index('Rank')
result
```

4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com

```python
url="https://bookpage.com/reviews"
page=requests.get(url)
soup=bs(page.content,'html.parser')
books=soup.find_all('div',class_="row-fluid article-row")
title=[]
author=[]
genre=[]
review=[]
for book in books:
    title.append(book.find('h4').get_text().replace("\n",""))
    author.append(book.find('p',class_="sans bold").get_text().replace("\n",""))
    genre.append(book.find('p',class_="genre-links hidden-phone").get_text().replace("\n",""))
    url_new=book.find('div',class_="read-full").find('a').get('href')
    page_new=requests.get("https://bookpage.com"+url_new)
    soup=bs(page_new.content,'html.parser')
    review.append(soup.find('div',class_="article-body").get_text().replace("\n",""))
data={'Title':title,'Author':author,'Genre':genre,'Review':review}
df=pd.DataFrame(data)
df
```

7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
product_name=[]
price=[]
image_url=[]
ratings=[]
for i in range(1,21):
    page= requests.get("https://www.amazon.in/s?k=mobile+under+20000&page="+str(i)+"&qid=1629979374&ref=sr_pg_"+str(i), headers={
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"})
    soup=bs(page.content,'html.parser')
    mobiles=soup.find_all("div",class_=["s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 AdHolder sg-col sg-col-12-of-16","s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16"])
    for mobile in mobiles:
        product_name.append(mobile.find("h2").get_text())
        price.append(mobile.find("span",class_="a-price-whole"))
        image_url.append(mobile.find("img",class_="s-image").get('src'))
        ratings.append(mobile.find('span',class_="a-icon-alt"))
Products={'Product Name':product_name,'Price':price,'Image URL':image_url,'Average Rating':ratings}
df=pd.DataFrame(data=Products)
df
```

8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
import datetime as dt
page=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YRzDFHUzZNg",headers={
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"})
soup=bs(page.content,'html.parser')
forecasts=soup.find_all('div',class_="row-forecast")
now=dt.datetime.now()
all_day_name = []
for i in range(0,6):
    all_day_name.append((now + dt.timedelta(days=i)).strftime("%A"))
data = []
for i, forecast in enumerate(forecasts):
    
    forecast_label = forecast.find('div',{'class':'forecast-label'}).text 
    forecast_text = forecast.find('div',{'class':'forecast-text'}).text 
    
    if i == 0:
        period = 'Today'
    elif forecast_label in all_day_name:
        period = forecast_label
    else:
        continue
        
    sort_desc = forecast_text.split('.')[0]
    temprature = forecast_text.split(', with a ')[1].split('.')[0][-2:]    
    description = forecast_text
    data.append({'period': period,'sort_desc': sort_desc,'temprature': temprature,'description': description})
        
df = pd.DataFrame(data) 
df
```

9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
title=[]
company=[]
ctc=[]
apply_date=[]
page= requests.get("https://internshala.com/fresher-jobs/page-1", headers={
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"})
soup=bs(page.content,'html.parser')
jobs=soup.find_all('div',class_="internship_meta")
for job in jobs:
    title.append(job.find("div",class_="heading_4_5 profile").get_text().replace("\n","").strip())
    company.append(job.find("div",class_="heading_6 company_name").get_text().replace("\n","").strip())
    x=job.find_all("div",{"class":"item_body","id":""})
    ctc.append(x[0].get_text().replace("\n","").strip())
    apply_date.append(x[1].get_text().replace("\n","").strip())
data={'Job Title':title,'Company':company,'CTC':ctc,'Apply Date':apply_date}
df=pd.DataFrame(data)
df
```
